```python
DF_Interfacility = pd.read_csv("Interfacility Orders anonymized.csv")
DF_ProductMaster = pd.read_csv("Product Master anonymized.csv")
```

### ProductMaster

```python
DF_ProductMaster.head()
```

```python
# Integrity Check 1: Check and remove duplicate ProductIDs
duplicate_count = DF_ProductMaster.duplicated(subset='DPPART').sum()
print(f"Count of duplicate ProductIDs before removal: {duplicate_count}")

DF_ProductMaster.drop_duplicates(subset='DPPART', keep='first', inplace=True)
```

```python
duplicate_count = DF_ProductMaster.duplicated(subset='DPPART').sum()
print(f"Count of duplicate ProductIDs after removal: {duplicate_count}")
```

```python
#Intergrity check 2:
# Identify purely text columns
text_columns = DF_ProductMaster.select_dtypes(include='object').columns

# Print the list of purely text columns
print("Purely Text Columns:")
print(text_columns)
```

```python
# List of purely text columns
text_columns = ['DPWH', 'DPPART', 'DPIUM', 'DPPUM', 'DPSUM', 'DPPLIN', 'DPPCLS',
                'DPDESC', 'DPBRND', 'DPPSTR', 'DPUSR3', 'DPVAS', 'DPCURC']

text_df = DF_ProductMaster[text_columns]

# Report out: % of blanks, # of different values, Top 10 most frequent values with % share for each text column
for column in text_columns:
    print(f"Column: {column}")
    print(f"Percentage of blanks: {text_df[column].isnull().mean() * 100:.2f}%")
    print(f"Number of different values: {text_df[column].nunique()}")
    print(f"Top 10 most frequent values with % share:")
    print(text_df[column].value_counts(normalize=True).head(10) * 100)
    print("\n")
```

```python
#Intergrity check 3

numerical_columns = DF_ProductMaster.select_dtypes(include='number').columns

# Print the list of numerical columns
print("Purely Numerical Columns:")
print(numerical_columns)
```

```python
# List of purely numerical columns
numerical_columns = ['DPMU', 'DPSAF', 'DPDIVN', 'DPCORG', 'DPWGTK', 'DPCOMV', 'DPCOGS']
numerical_df = DF_ProductMaster[numerical_columns]

# Report out: % of blanks, Minimum, Average, Median, Maximum for each numerical column
for column in numerical_columns:
    print(f"Column: {column}")
    print(f"Percentage of blanks: {numerical_df[column].isnull().mean() * 100:.2f}%")
    print(f"Minimum: {numerical_df[column].min()}")
    print(f"Average: {numerical_df[column].mean()}")
    print(f"Median: {numerical_df[column].median()}")
    print(f"Maximum: {numerical_df[column].max()}")
    print("\n")
```
